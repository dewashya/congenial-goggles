```python
col_dtype = list(nav.dtypes)
col_dtest=set(col_dtype)
col_dtest
#so we still have dtype object in our set.
```

```python
o = 0
noo = list()
for d in col_dtype:
    if type(d) == type(col_dtype[90]):
        noo.append(o)
    o += 1
```

```python
#now i have the index of all the columns that had object datatype. Lets fix them.
```

```python
needc = nav.iloc[:,noo]
```

```python
needc
```

```python
correct=list()
for n in needc :
    l = list(needc[n].unique())
    for ll in l:
        if type(ll) != type(float()):
            if type(ll) != type(int()):
                if ll in correct:
                    continue
                else:
                    correct.append(ll)
```

```python
print(correct)
```

```python
nav
```

```python
nav.replace(correct,float("nan") ,inplace = True)
```

```python
col_dtype = list(nav.dtypes)
col_dtest=set(col_dtype)
col_dtest
```
